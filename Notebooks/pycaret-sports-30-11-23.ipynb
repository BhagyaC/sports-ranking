# Setting up the regression data

```python
exp = exp.setup(df, target = 'score', session_id = 123, remove_outliers = True, normalize = True)
```

```python
df.dtypes
```

```python
models = exp.compare_models()
```

```python
rf = exp.create_model('rf')
```

```python
tuned_rf = exp.tune_model(rf)
```

# Stacking the top performing model

```python
stacker = exp.stack_models(estimator_list = exp.compare_models(n_select=5, fold = 5, include = exp.models(type='ensemble').index.tolist()))
```
